#### Word-Cloud on Positive Polarity

```python
#clean the text with Positive polarity
positive_text = data_sample[data_sample['sentiment']==1]['text']
```

```python
#start time
st = time.time()
positive_text_processed = positive_text.apply(lambda x:text_preprocessing(x,remove_nonalphanumeric=1,remove_stopword=1
                                                                         ,remove_urls=1,lemmatization=0,len_word_remove_flag=1))
#end time
et = time.time()
print("Time to pre-process all positive texts without lemmatizing:",et-st)
positive_text_processed = positive_text_processed.reset_index(drop=True)
```

```python
pos_word_counts = {}
for index in tqdm(range(0,len(positive_text_processed))):
    comment = positive_text_processed[index]
    tokens = comment.split(' ')
    
    for token in tokens:
        if token in pos_word_counts:
            pos_word_counts[token] = pos_word_counts[token]+1
        else:
            pos_word_counts[token]=1
```

```python
# Generate word cloud
#plt.figure(figsize = (5,5))
wordcloud = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(pos_word_counts)

# Display the generated image
plt.figure(figsize=(5, 5), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

```python
postive_comments_word_freq = pd.DataFrame({'Word':pos_word_counts.keys(),'Freq':pos_word_counts.values()})
postive_comments_word_freq = postive_comments_word_freq.sort_values('Freq',ascending=False)
postive_comments_word_freq = postive_comments_word_freq.reset_index(drop=True)

#Filter the top-10 words
postive_comments_word_freq_top10 = postive_comments_word_freq.iloc[0:9,:]

plt.figure(figsize = (6,5))

# Create a bar plot of the top words
x = list(postive_comments_word_freq_top10['Word'])
y = list(postive_comments_word_freq_top10['Freq'])
plt.bar(x, y)

# Add labels and title
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Words in Positive Comment')

# Display the plot
plt.show()
```

#### words coming frequently across both Positive and Negative comments to be cleaned

```python
#There are words which are common across both Positive and Negative comments 
#Such word do not add any weightage to the final output and can be cleaned
positive_words_dict = dict(pos_word_counts)
Positive_comments_df = pd.DataFrame({'Positive_Words':list(positive_words_dict.keys()),'Positive_Frequency':list(positive_words_dict.values())})
Positive_comments_df = Positive_comments_df.sort_values('Positive_Frequency', ascending=False)
Positive_comments_df = Positive_comments_df.reset_index(drop=True)
cumulative_positive_count = Positive_comments_df['Positive_Frequency'].cumsum()
# add 'Cumulative Sales' column to DataFrame
Positive_comments_df['cumulative_positive_count'] = cumulative_positive_count
Positive_comments_df['cumulative_percentage_mix'] = Positive_comments_df['cumulative_positive_count']/Positive_comments_df['Positive_Frequency'].sum()

#Limit to select the vocabulary which make up for 80% of the total word counts
print("Total number of words in Positive comments:",Positive_comments_df.shape[0])
Positive_comments_df = Positive_comments_df[Positive_comments_df['cumulative_percentage_mix']<0.80]
print("Total number of words making of total 80% of word counts:",Positive_comments_df.shape[0])

negative_words_dict = dict(neg_word_counts)
Negative_comments_df = pd.DataFrame({'Negative_Words':list(negative_words_dict.keys()),'Negative_Frequency':list(negative_words_dict.values())})
Negative_comments_df = Negative_comments_df.sort_values('Negative_Frequency', ascending=False)
Negative_comments_df = Negative_comments_df.reset_index(drop=True)
cumulative_negative_count = Negative_comments_df['Negative_Frequency'].cumsum()
# add 'Cumulative Sales' column to DataFrame
Negative_comments_df['cumulative_negative_count'] = cumulative_negative_count
Negative_comments_df['cumulative_percentage_mix'] = Negative_comments_df['cumulative_negative_count']/Negative_comments_df['Negative_Frequency'].sum()

#Limit to select the vocabulary which make up for 80% of the total word counts
print("Total number of words in Negative comments:",Negative_comments_df.shape[0])
Negative_comments_df = Negative_comments_df[Negative_comments_df['cumulative_percentage_mix']<0.80]
print("Total number of words making of total 80% of word counts:",Negative_comments_df.shape[0])

```

```python
Positive_comments_df.head()
```

```python
#Select the unique list of words in Positive and Negative comments
positive_words_list = list(Positive_comments_df['Positive_Words'])
negative_words_list = list(Negative_comments_df['Negative_Words'])

word_list = list()
word_list.extend(positive_words_list)
word_list.extend(negative_words_list)
word_list = list(set(list(word_list)))
word_list = [word for word in word_list if len(word)>1]
```

```python
word_lift_df = pd.DataFrame({'Word':word_list})
word_lift_df = pd.merge(word_lift_df,Positive_comments_df.loc[:,['Positive_Words','Positive_Frequency']]
         ,left_on='Word',right_on='Positive_Words',how='left')

word_lift_df = pd.merge(word_lift_df,Negative_comments_df.loc[:,['Negative_Words','Negative_Frequency']]
         ,left_on='Word',right_on='Negative_Words',how='left')

word_lift_df['Positive_Frequency']=word_lift_df['Positive_Frequency'].fillna(0)
word_lift_df['Negative_Frequency']=word_lift_df['Negative_Frequency'].fillna(0)

word_lift_df['Negative_Words'] = np.where(word_lift_df['Negative_Words'].isna(),word_lift_df['Positive_Words'],word_lift_df['Negative_Words'])
word_lift_df['Positive_Words'] = np.where(word_lift_df['Positive_Words'].isna(),word_lift_df['Negative_Words'],word_lift_df['Positive_Words'])
```

```python
word_lift_df['pos_lift']=(word_lift_df['Positive_Frequency']-word_lift_df['Negative_Frequency'])/np.where(word_lift_df['Positive_Frequency']==0,1,word_lift_df['Positive_Frequency'])
```

```python
word_lift_df['neg_lift']=(word_lift_df['Negative_Frequency']-word_lift_df['Positive_Frequency'])/np.where(word_lift_df['Negative_Frequency']==0,1,word_lift_df['Negative_Frequency'])
```

```python
word_lift_df.sort_values('pos_lift',ascending=False)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a dictionary with word frequency
word_lift_df_filtered = word_lift_df[word_lift_df['Positive_Frequency']>500]

word_counts = {}
for word,value in zip(word_lift_df_filtered['Positive_Words'],word_lift_df_filtered['pos_lift']):
    word_counts[word] = value

# Generate word cloud
wordcloud = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(word_counts)

# Display the generated image
plt.figure(figsize=(5, 5), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

```python
# Create a dictionary with word frequency

word_lift_df_filtered = word_lift_df[word_lift_df['Negative_Frequency']>500]
word_counts = {}
for word,value in zip(word_lift_df_filtered['Negative_Words'],word_lift_df_filtered['neg_lift']):
    word_counts[word] = value

# Generate word cloud
wordcloud = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(word_counts)

# Display the generated image
plt.figure(figsize=(5, 5), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)

plt.show()
```

### Model Building

#### Model 1 - GBM classifier with TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
#Split the data into Train and Test to avoid inducing any bias encountered during preprocessing
X = data_sample['text']
ylabels = data_sample['sentiment']

x_train,x_test,y_train,y_test = train_test_split(X,ylabels,test_size=0.2,stratify = ylabels)

print("X train size:",x_train.shape)
print("X test size:",x_test.shape)
```

```python
data_sample.head()
```

```python
x_train = x_train.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)

x_test = x_test.reset_index(drop=True)
y_test = y_test.reset_index(drop=True)
```

#### Pre-Processing Steps for 1st iteration
* Removed links-URLs
* Removed nonalphanumeric words

```python
#pre-process the dataset
st = time.time()
x_train_cleaned = x_train.apply(lambda x:text_preprocessing(x,remove_nonalphanumeric=1,remove_stopword=0
                                                                         ,remove_urls=1,lemmatization=0,len_word_remove_flag=0))
#end time
et = time.time()
print("Time to pre-process training data - Iteration1:",et-st)
```

```python
#computing the model metrics with TF-IDF vectorizer
#ngram_range = ngrams to be considered
#min_df = ignore terms that have a document frequency lower than given threshold

tfidf_vector = TfidfVectorizer(ngram_range =(1,2),min_df=20)
x_train_vectors_tfidf = tfidf_vector.fit_transform(x_train_cleaned)

#Shape of datasets after transformation
print("Train tfidf vectors shape:",x_train_vectors_tfidf.shape)
#x_test_vectors_tfidf = tfidf_vector.transform(x_test)
```

```python
# Set Gradient Boosting parameters
# using the base estimators for first iteration
params = {
    'n_estimators': 100,
    'learning_rate': 0.1,
    'max_depth': 3,
    'subsample': 0.8,
    'min_samples_split': 2,
    'min_samples_leaf': 1,
    'random_state': 42
}
```

```python
model_gbm1 = GradientBoostingClassifier(**params)
model_gbm1.fit(x_train_vectors_tfidf, y_train)
```

```python
# Make predictions on the test set and calculate accuracy
x_test_cleaned = x_test.apply(lambda x:text_preprocessing(x,remove_nonalphanumeric=1,remove_stopword=0
                                                                         ,remove_urls=1,lemmatization=0,len_word_remove_flag=0))
x_test_vectors_tfidf = tfidf_vector.transform(x_test_cleaned)


y_pred_label = model_gbm1.predict(x_test_vectors_tfidf)
y_pred_prob = model_gbm1.predict_proba(x_test_vectors_tfidf)
```

```python
test_dataframe_output = pd.DataFrame({'Label':y_test
                                      ,'Predicted_label':list(y_pred_label)
                                      ,'Predicted_proba_negative':list(y_pred_prob[:,0])
                                      ,'Predicted_proba_positive':list(y_pred_prob[:,1])})
```

```python
#confusion matrix
cm = confusion_matrix(test_dataframe_output['Label'], test_dataframe_output['Predicted_label'])

# get the class labels
labels = np.unique(test_dataframe_output['Label'])

# create a dataframe with row and column labels
cm_df = pd.DataFrame(cm, index=labels, columns=labels)

# print the dataframe
print(cm_df)
print("Accuracy of the model:",metrics.accuracy_score(test_dataframe_output['Label'],test_dataframe_output['Predicted_label']))
print("Precision of the model:",metrics.precision_score(test_dataframe_output['Label'],test_dataframe_output['Predicted_label']))
print("Recall of the model:",metrics.recall_score(test_dataframe_output['Label'],test_dataframe_output['Predicted_label']))
print("-------------------Classification Report----------------------")
print(metrics.classification_report(test_dataframe_output['Label'],test_dataframe_output['Predicted_label']))
```

```python
roc_auc,fpr,tpr,thresholds = auc_roc(test_dataframe_output['Label'],test_dataframe_output['Predicted_proba_positive'])
```

```python
# Plot the AUC curve
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC for GBM Base Model')
plt.legend(loc="lower right")
plt.show()
```
